```python
import torchaudio
from speechbrain.inference.speaker import EncoderClassifier
classifier = EncoderClassifier.from_hparams(source="speechbrain/spkrec-ecapa-voxceleb")
signal, fs =torchaudio.load("/content/Sample_data_voices/000950s_Female_shout_uk.wav")
embeddings = classifier.encode_batch(signal)
```

```python
from speechbrain.inference.speaker import SpeakerRecognition
verification = SpeakerRecognition.from_hparams(source="speechbrain/spkrec-ecapa-voxceleb", savedir="pretrained_models/spkrec-ecapa-voxceleb")
score, prediction = verification.verify_files("/content/Sample_data_voices/000950s_Female_shout_uk.wav", "/content/Sample_data_voices/000950s_Female_shout_uk.wav") # Different Speakers
#score, prediction = verification.verify_files("tests/samples/ASR/spk1_snt1.wav", "tests/samples/ASR/spk1_snt2.wav") # Same Speaker
print(score, prediction)
```
